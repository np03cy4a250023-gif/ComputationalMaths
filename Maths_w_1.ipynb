Write a function fib (n) that prompts the number of terms “n” from the user and
displays first “n” terms of Fibonacci sequence iteratively.

```python
def fib(n):
  i=0
  a=0
  b=1
  n=n
  while i<n:
    print(a)
    c=a+b
    a=b
    b=c
    i=i+1
fib(7)
```

Write a function SumDigits(n) that displays the sum of digits of “n”. Hint:
SumDigits(234) will be 2+3+4=9

```python
def SumDigits(n):
    total = 0
    while n > 0:
        digit = n % 10
        total += digit
        n = n // 10
    print(total)

SumDigits(234)

```

Write a function squaredSum that returns the sum of the squares of an array/list of integers such that squaredSum ([2,3,4]) gives 29. Write your solution in pure Python to show the logic in your solution, do not use any library functions.

```python
def squaredSum(arr):
    total = 0

    for num in arr:
        total = total + (num * num)

    return total


print(squaredSum([2, 3, 4]))
```
