**Problem 2: Given a matrix and a vector, write a Python function that performs Matrix-Vector
product. Check your result with NumPy built-in function to perform Matrix-Vector/Matrix-
Matrix product.**

```python
import numpy as np
def dot_product(vec1,vec2):
  #Flatten the input vectors
  vec1 = vec1.flatten()
  vec2=vec2.flatten()
  return np.sum(np.multiply(vec1,vec2))

def matrix_vector_product(mat, vec):
  m,n= mat.shape
  p,q= vec.shape

  assert n == p, "the matrix and vector are not compatible for multiplication!"

  #create a ndarray to store the result
  result=np.zeros((m,q),dtype=float)

  #loop to perform Matrix-vector product
  #perform dot product between the vector and each row of the matrix

  for i in range(m):
    result[i]=dot_product(mat[i,:],vec)

  return result


if __name__ == "__main__":

  """Part 2: Problem 2: Given a matrix and a vector, write a function that performs Matrix-vector product."""
  mat1= np.full((3,3),3,dtype=int)
  vec1= np.full((3,1),2,dtype=int)
  print("Matrix =")
  print(mat1)
  print("vector =")
  print(vec1)

  #call the function matrix_vector_product_(mat,vec)

  result1=matrix_vector_product(mat1,vec1)
  print("Result =")
  print(result1)

  #check with numpy in-built function for multiplication
  result2=mat1 @ vec1
  print("Result =")
  print(result2)
```

**Problem 3: Given two compatible matrices, write a Python function that performs Matrix-Matrix
product. Check your result with NumPy built-in function to perform Matrix-Vector/Matrix-
Matrix product.**

```python
import numpy as np

def matrix_multiply(A, B):
    # Check if multiplication is possible
    if A.shape[1] != B.shape[0]:
        print("Matrix multiplication not possible!")
        return None

    # Result matrix (m x p)
    C = np.zeros((A.shape[0], B.shape[1]))

    # Triple nested loop
    for i in range(A.shape[0]):        # rows of A
        for j in range(B.shape[1]):    # columns of B
            for k in range(A.shape[1]):  # shared dimension
                C[i][j] += A[i][k] * B[k][j]

    return C
    A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

# Manual multiplication
result_manual = matrix_multiply(A, B)
print("Manual Result:\n", result_manual)

# NumPy built-in multiplication
result_numpy = np.dot(A, B)
print("\nNumPy Result:\n", result_numpy)
```

**Problem 4: Given a matrix, write a Python function that checks if it is a symmetric matrix.**

**Hint: Use the definition based on Transpose of Matrix.**

```python
import numpy as np

def is_symmetric(A):
    # A matrix must be square first
    if A.shape[0] != A.shape[1]:
        return False

    n = A.shape[0]

    # Check element by element
    for i in range(n):
        for j in range(n):
            if A[i][j] != A[j][i]:
                return False

    return True
    A = np.array([
    [1, 2, 3],
    [2, 5, 6],
    [3, 6, 9]
])

B = np.array([
    [1, 0, 0],
    [2, 5, 0],
    [3, 6, 9]
])

print("A is symmetric:", is_symmetric(A))
print("B is symmetric:", is_symmetric(B))
```
