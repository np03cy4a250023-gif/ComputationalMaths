**Example 1: We will start with a simple problem. Consider the following linear systems:**

**(a) x+3y=5**

**𝑥−2𝑦=−5**

**𝑥+𝑦=3**

```python
import numpy as np

import matplotlib.pyplot as plt

""" Part 1 - Problem 1: Solve linear systems graphically. """

# Generate 100 equidistant point between -5 to 5 for x

x = np.linspace(-5,5,100)

## Calculate y values for Equation 1 and 2

y1 = (5-x)/3

y2 = (5+x)/2

y3 = 3-x

## Generate a figure frame for plotting the lines

fig, ax = plt.subplots()

## Plot the two lines

ax.plot(x,y1, color='r')

ax.plot(x,y2, color='g')

ax.plot(x,y3, color='b')

## To decorate the plot, we insert equations.

ax.text(0.5,0.6, '$x+3y = 5$', color='r')

ax.text(-3,0.5,'$x-2y = -5$', color='g')

ax.text(1.2,2, '$x+y = 3$', color='b')

## Set the limits of the X & Y axes.

ax.set_xlim(-4,4)

ax.set_ylim(-2,6)

## Set property of X and Y axes, e.g., color and linewidth

ax.axvline(color='k',linewidth = 1)

ax.axhline(color='k',linewidth = 1)

## This option specifies the ticks based the list of numbers provided.

ax.set_xticks(list(range(-4,5)))

ax.set_aspect('equal')

ax.grid(True,ls=':')

plt.xlabel("X")

plt.ylabel("Y")

plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

""" Part 1 - Problem 1: Solve linear systems graphically. """

# Generate 100 equidistant point between -5 to 5 for x

x = np.linspace(-5,5,100)

## Calculate y values for Equation 1, 2 and 3

y1 = (3-2*x)/3

y2 = (6-4*x)/6

y3 = (3*x-1)/2

## Generate a figure frame for plotting the lines

fig, ax = plt.subplots()

## Plot the two lines

ax.plot(x,y1, color='r')

ax.plot(x,y2, color='g')

ax.plot(x,y3, color='b')

## To decorate the plot, we insert equations.

ax.text(0.5,0.6, '$2x+3y = 3$', color='r')

ax.text(-3,0.5,'$4x+6y = 6$', color='g')

ax.text(1.2,2, '$3x-2y = 1$', color='b')

## Set the limits of the X & Y axes.

ax.set_xlim(-4,4)

ax.set_ylim(-2,6)

## Set property of X and Y axes, e.g., color and linewidth

ax.axvline(color='k',linewidth = 1)

ax.axhline(color='k',linewidth = 1)

## This option specifies the ticks based the list of numbers provided.

ax.set_xticks(list(range(-4,5)))

ax.set_aspect('equal')

ax.grid(True,ls=':')

plt.xlabel("X")

plt.ylabel("Y")

plt.show()
```

**1. Create a matrix 𝑨 of order 5 × 5 with random entries. Calculate the rank of the matrix 𝑨 to check the number of linearly independent columns or rows and to check if the matrix is invertible. What are the other ways to check if the matrix is invertible?**

```python
import numpy as np

A = np.random.randint(1,11,(5,5))
r = np.linalg.matrix_rank(A)

print(A)
print("Rank:", r)

if r == 5:
    print("Invertible")
else:
    print("Not Invertible")
```

**2. Generate a matrix 𝑨 of order 5 × 5 with random entries and generate a vector (column) 𝒙 of length 5 with random entries. Calculate Matrix-vector product 𝒃 = 𝑨𝒙. Now, we have a linear system, i.e. the matrix 𝑨 and a vector 𝒃. Solve this Linear System.**



```python
import numpy as np

A = np.random.randint(1,11,(5,5))
x = np.random.randint(1,11,(5,1))

b = A @ x

print("A:\n", A)
print("x:\n", x)
print("b:\n", b)

# Solve Ax = b
try:
    sol = np.linalg.solve(A, b)
    print("Solution x:\n", sol)
except:
    print("Matrix A is not invertible")
```

**1. Consider the following 10 points in X-Y plane: (−3, −7.6), (−2,1), (−1, 4.8), (0, 5), (1, 2.8), (2, −0.6), (3, −4), (4, −6.2), (5, −6), (7, 6.4). Plot these points in X-Y plane. Connect the points to guess the polynomial. Once you guess your polynomial, formulate a Linear System and then solve it to find the coefficient of the polynomial. Finally, check if the polynomial fits along with the above given points.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
x = np.array([-3,-2,-1,0,1,2,3,4,5,7], dtype=float)
y = np.array([-7.6,1,4.8,5,2.8,-0.6,-4,-6.2,-6,6.4], dtype=float)

# Use 4th degree polynomial (5 points system)
x1 = np.array([-2,-1,0,1,2], dtype=float)
y1 = np.array([1,4.8,5,2.8,-0.6], dtype=float)

# Matrix A
A = np.array([[xi**4, xi**3, xi**2, xi, 1] for xi in x1])

# Solve for coefficients
a, b, c, d, e = np.linalg.solve(A, y1)

# Polynomial function
def f(x):
    return a*x**4 + b*x**3 + c*x**2 + d*x + e

# Check values
for xi, yi in zip(x, y):
    print(xi, yi, round(f(xi),2))

# Plot
xp = np.linspace(-3,7,200)
plt.scatter(x, y, color='red')
plt.plot(xp, f(xp), color='blue')
plt.grid()
plt.show()
```
