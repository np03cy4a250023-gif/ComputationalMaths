```python
import matplotlib.pyplot as plt

#create a function to calculate y values

def function(x):

  result= 3*x+2

  return result

#defining empty lists

xCoord=[]

yCoord=[]

#loop over 10 values from 0 to 9, and calculate the matching values for #y using the function


for i in range(0,10):

  xCoord.append(i)
  yCoord.append(function(i))



plt.plot(xCoord,yCoord, marker='x',color='red')

#function to add title

plt.title('Example 5')


#Function to label axes plt.xlabel('X-axis')

plt.ylabel('Y-axis')


# function for the visibility of grid plt.grid()

plt.show()
```

```python
# Import necessary modules

# Using math module for pi and sin function


import matplotlib.pyplot as plt

import math


# Step 1: Generate equidistant x-coordinates between the given interval

# Lists for coordinates

x = []

y = []


# Define start, end, and step size for x increment

start = -4 * math.pi

end = 4 * math.pi

step = 0.1 # Controls smoothness of the graph


i = start

while i <= end:

  x.append(i) # Store x-coordinate

# Step 2: Calculate the y-coordinate using sin function for each x

y.append(math.sin(i))

i += step


# Step 3: Plot the graph using calculated coordinates

# Label the graph properly

plt.plot(x, y, color="red", label="sin function")

plt.xlabel("X")

plt.ylabel("Y")

plt.title("y = sin(x) ,-4π<= x <= 4π")

plt.grid(True)

plt.legend()

plt.show()
```
