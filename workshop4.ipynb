Task 1: Create a scatter graph of 10 random x and y values between 0 and 50. Label the axis to show
x – Time taken in seconds
y – Distance reached
Title “Buzzer Game Results”
One you have a working program, add the following code to your program;
colors=["red","green","blue","yellow","pink","orange","purple","brown"
,"cyan","magenta"]
And add to your scatter() call the ‘c=colors’ argument like so;
plt.scatter(x, y, c=colors)
What happens?

```python
import matplotlib.pyplot as plt
import random

#Generate 10 random values between 0 and 50
x=[random.randint(0,50) for _ in range(10)]
y=[random.randint(0,50) for _ in range(10)]

colors = ["red","green","blue","yellow","pink","orange","purple","brown",
"cyan","magenta"]

#Scatter plot
plt.scatter(x,y, c=colors)

#Labels and title
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game results")

plt.show()
```

```python
#a)
import matplotlib.pyplot as plt

# Data
years = [2017, 2018, 2019, 2020, 2021, 2022]
bears = [8, 54, 93, 116, 137, 184]
dolphins = [150, 77, 32, 11, 6, 1]
whales = [80, 54, 100, 76, 93, 72]

labels = ["Bears", "Dolphins", "Whales"]

# Loop to create pie chart for each year
for i in range(len(years)):
    data = [bears[i], dolphins[i], whales[i]]

    plt.figure()  # new figure for each year
    plt.pie(data, labels=labels, autopct='%1.1f%%')
    plt.title(f"Animal Population in {years[i]}")

    plt.show()
```

```python
#b)
import matplotlib.pyplot as plt

years = [2017, 2018, 2019, 2020, 2021, 2022]
bears = [8, 54, 93, 116, 137, 184]
dolphins = [150, 77, 32, 11, 6, 1]
whales = [80, 54, 100, 76, 93, 72]

labels = ["Bears", "Dolphins", "Whales"]

# Create subplots (2 rows, 3 columns)
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

axes = axes.flatten()  # make it easier to loop

for i in range(len(years)):
    data = [bears[i], dolphins[i], whales[i]]

    axes[i].pie(data, labels=labels, autopct='%1.1f%%')
    axes[i].set_title(str(years[i]))

plt.tight_layout()
plt.show()
```

Task 3: Construct a pie chart to visually display the favorite fruits of the students in a class based on the given data: Mango - 45; Orange - 30; Plum - 15; Pineapple - 30; Melon – 30

```python
import matplotlib.pyplot as plt

# Data
fruits = ["Mango", "Orange", "Plum", "Pineapple", "Melon"]
values = [45, 30, 15, 30, 30]

# Create pie chart
plt.pie(values, labels=fruits, autopct='%1.1f%%')

# Title
plt.title("Favorite Fruits of Students")

# Show chart
plt.show()
```

```python
#a)
import matplotlib.pyplot as plt

# Data
cars = ["Red", "Silver", "Blue", "White", "Green"]
values = [123, 230, 78, 193, 12]

# Bar chart
plt.bar(cars, values)

plt.xlabel("Car Color")
plt.ylabel("Number Seen")
plt.title("Car Color Count")

plt.show()
```

```python
#b)
plt.bar(cars, values, width=0.3)
plt.show()
```

```python
#c)
plt.barh(cars, values)
plt.show()
```

```python
#d)
colors = ["red", "silver", "blue", "white", "green"]

plt.bar(cars, values, color=colors)
plt.title("Car Color Count")
plt.show()
```

```python
#e)
plt.pie(values, labels=cars, autopct='%1.1f%%')
plt.title("Car Color Distribution")
plt.show()
```

```python
#f)
myexplode = [0.3, 0, 0, 0, 0]

plt.pie(values, labels=cars, explode=myexplode, autopct='%1.1f%%')
plt.title("Car Color Distribution")
plt.show()
```
