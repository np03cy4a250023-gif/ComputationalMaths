**a. Prepare X and Y coordinates values for the function f(x) = 4x2+2 using NumPy array and plot the graph using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Create X values
x = np.linspace(-10, 10, 100)

# Define function f(x) = 4x^2 + 2
y = 4 * x**2 + 2

# Plot graph
plt.plot(x, y,marker ='o',label="f(x) = 4x^2 + 2")

# Add labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Graph of f(x) = 4x^2 + 2")

# Show grid and legend
plt.grid(True)
plt.legend()

# Display plot
plt.show()
```

**Create two vectors in 2-dimensional (2D) space and plot them using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
""" Problem 2: Create 2-dimensional position vectors and plot them. """
origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()
# Add vectors V and W to the plot
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

**a. Create three vectors in 3D space and plot them using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vectors using quiver (arrows)
ax.quiver(0, 0, 0, U[0], U[1], U[2], color='r', label='U')
ax.quiver(0, 0, 0, V[0], V[1], V[2], color='g', label='V')
ax.quiver(0, 0, 0, W[0], W[1], W[2], color='b', label='W')

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Set title
ax.set_title("3D Vector Plot")

# Show plot
plt.show()
```

**b. Create two vectors in 2D space, and add them to create a third vector, and plot all three vectors using Matplotlib.**

**U = np.array([-0.5, 0.5])**

**V = np.array([1,1])**

**The two vectors are added as**

**W = U + V # adding two vector**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
U = np.array([-0.5, 0.5])
V = np.array([1, 1])

# Vector addition
W = U + V

# Create plot
plt.figure()

# Plot vectors using arrows (quiver)
plt.quiver(0, 0, U[0], U[1], angles='xy', scale_units='xy', scale=1, color='r', label='U')
plt.quiver(0, 0, V[0], V[1], angles='xy', scale_units='xy', scale=1, color='g', label='V')
plt.quiver(0, 0, W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b', label='W = U + V')

# Set limits
plt.xlim(-1, 2)
plt.ylim(-1, 2)

# Grid and labels
plt.grid(True)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("2D Vector Addition")

# Show plot
plt.show()
```

**c. Write a function to calculate magnitude of a vector. For example, calculate the magnitude of the vector be U = (1.5, 1.5, 2.0)**

```python
import numpy as np

def magnitude(v):
    return np.sqrt(np.sum(v**2))

U = np.array([1.5, 1.5, 2.0])

print("Magnitude of U:", magnitude(U))
```

**d. Write a function to calculate the direction cosines of the vector.
For example, let the vector be O = (1.5, 1.5, 2.0).**

```python
def direction_cosines(v):
    mag = np.sqrt(np.sum(v**2))
    return v / mag

O = np.array([1.5, 1.5, 2.0])

print("Direction cosines of O:", direction_cosines(O))
```

**e. Write a function to calculate Scalar (dot) product of two vectors. For example, let the vectors be a= (1.5, 1.5, 2.0) and b = (−1.5, 0.5, 2.0)**

```python
def dot_product(a, b):
    return np.sum(a * b)

a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])

print("Dot product:", dot_product(a, b))
```

**f. Write a function to calculate Vector (cross) product of two vectors. For example, let the vectors
be a = (1.5, 1.5, 2.0) and b= (−1.5, 0.5, 2.0).**

```python
def cross_product(a, b):
    return np.cross(a, b)

a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])

print("Cross product:", cross_product(a, b))
```
